## **Conjunto de dados:**

**Marca:** Especifica a marca do carro (por exemplo, Toyota, BMW, Ford).
**Valores de exemplo:** "Toyota", "BMW", "Mercedes".

**Modelo:** Especifica o modelo do carro (por exemplo, Corolla, Focus, X5).
**Valores de exemplo:** "Corola", "Foco", "X5".

**Ano:** O ano de produ√ß√£o do carro. Os anos mais recentes geralmente indicam pre√ßos mais altos.
**Valores de exemplo:** 2005, 2018, 2023.

**Engine_Size:** Especifica o tamanho do motor em litros (L). Motores maiores geralmente se correlacionam com pre√ßos mais altos.
**Valores de exemplo:** 1.6, 2.0, 3.5.

**Fuel_Type:** indica o tipo de combust√≠vel utilizado pelo carro:
**Gasolina:** Carros movidos a gasolina.
**Diesel:** Carros movidos a √≥leo diesel.
**H√≠brido:** Carros que usam combust√≠vel e eletricidade.
**El√©trico:** Carros totalmente el√©tricos.

**Transmiss√£o:** O tipo de transmiss√£o no carro:
**Manual:** Transmiss√£o manual.
**Autom√°tico:** Transmiss√£o autom√°tica.
**Semi-Autom√°tico:** Transmiss√£o semi-autom√°tica.

**Quilometragem:** A dist√¢ncia total percorrida pelo carro, medida em quil√¥metros. Quilometragem mais baixa geralmente indica um pre√ßo mais alto.
**Valores de exemplo:** 15.000, 75.000, 230.000.

**Portas:** O n√∫mero de portas no carro. Geralmente 2, 3, 4 ou 5 portas.
Valores de exemplo: 2, 3, 4, 5.

**Owner_Count:** O n√∫mero de propriet√°rios anteriores do carro. Menos propriet√°rios geralmente indicam um pre√ßo mais alto.
Valores de exemplo: 1, 2, 3, 4.

**Pre√ßo:** O pre√ßo de venda estimado do carro. √â calculado com base em v√°rios fatores, como ano de produ√ß√£o, tamanho do motor, quilometragem, tipo de combust√≠vel e transmiss√£o.

**Passo 3 - Corre√ß√£o do DataFrame (Limpeza, organiza√ß√£o e modelagem)**

---

```python
# Fun√ß√£o para substituir os valores das colunas Tipo_Combustivel e Transmissao de texto 
# para n√∫meros e vice-versa
def codificar():
    if (df_carro["Tipo_Combustivel"].dtypes == object) and (df_carro["Transmissao"].dtypes == object):
        # Codificar de texto para n√∫mero
        df_carro["Tipo_Combustivel"] = df_carro["Tipo_Combustivel"].replace({
            "Diesel": 0,
            "Electric": 1,
            "Hybrid": 2,
            "Petrol": 3                         
        })
        
        df_carro["Transmissao"] = df_carro["Transmissao"].replace({
            "Automatic": 0,
            "Manual": 1,
            "Semi-Automatic": 2
        })
    else:
        # Cofificar de n√∫mero para texto
        df_carro["Tipo_Combustivel"] = df_carro["Tipo_Combustivel"].replace({
            0: "Diesel",
            1: "Electric",
            2: "Hybrid",
            3: "Petrol"                          
        })
        
        df_carro["Transmissao"] = df_carro["Transmissao"].replace({
            0: "Automatic",
            1: "Manual",
            2: "Semi-Automatic"
        })

```

```python
# Altera tipo da coluna Preco para float
df_carro[["Preco"]] = df_carro[["Preco"]].astype(float)

display(df_carro.dtypes)

display(df_carro)
```

Dataframe corrigido e remodelado.

```python
# Importando as bibliotecas necess√°rias
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Criando um novo dataframe para a IA, a partir do dataframe original
df_carro_ia = df_carro

display(df_carro_ia)
```

---

**Passo 1 - Preparar o dataframe para a Intelig√™ncia Artifical**

---

```python
# Codificar as colunas para tipo num√©rico para a Intelig√™ncia Artificial

# Codificar coluna Marca
codificador_marca = LabelEncoder()
df_carro_ia["Marca"] = codificador_marca.fit_transform(df_carro_ia["Marca"])

# Codificar coluna Tipo Modelo
codificador_modelo = LabelEncoder()
df_carro_ia["Modelo"] = codificador_modelo.fit_transform(df_carro_ia["Modelo"])

# Codificar coluna Tipo_Combustivel
codificador_combustivel = LabelEncoder()
df_carro_ia["Tipo_Combustivel"] = codificador_combustivel.fit_transform(df_carro_ia["Tipo_Combustivel"])

# Codificar coluna Transmissao
codificador_transmissao = LabelEncoder()
df_carro_ia["Transmissao"] = codificador_transmissao.fit_transform(df_carro_ia["Transmissao"])

df_carro_ia.info()

display(df_carro_ia)
```

```python
# Preparar o dataframe para o treino

# y √© a coluna do dataframe a ser prevista
y = df_carro_ia["Preco"]

# x s√£o as colunas do dataframe que ser√£o usadas para a previs√£o
x = df_carro_ia.drop(columns=["Preco"])

# Separar os dados de treino e os dados de teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)
```

---

**Passo 2 - Treinar a Intelig√™ncia Artificial**

---

```python
# Criar o medelo -> Prever os Pre√ßos em USD

# Arvore de Decis√£o -> RandonForest
# Nearest Neighbors -> KNN -> KNeighbors
# Regress√£o Linear -> RLN -> LinearRegression

# Cria a IA
modelo_arvore_decisao = RandomForestRegressor()
modelo_knn = KNeighborsRegressor()
modelo_rln = LinearRegression()

# Treinar a IA
modelo_arvore_decisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
modelo_rln.fit(x_treino, y_treino)
```

---

**Passo 3 - Escolher o melhor modelo**

---

```python
# Testar os modelos

# Testar os modelos
previsao_arvore_decisao = modelo_arvore_decisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)
previsao_rln = modelo_rln.predict(x_teste)

# Calcular os erros
print(f'MAE √Årvore de Decis√£o: {mean_absolute_error(y_teste, previsao_arvore_decisao):,.2f}')
print(f'MAE KNN: {mean_absolute_error(y_teste, previsao_knn):,.2f}')
print(f'MAE Regress√£o Linear: {mean_absolute_error(y_teste, previsao_rln):,.2f}')

print(f'MSE √Årvore de Decis√£o: {mean_squared_error(y_teste, previsao_arvore_decisao):,.2f}')
print(f'MSE KNN: {mean_squared_error(y_teste, previsao_knn):,.2f}')
print(f'MSE Regress√£o Linear: {mean_squared_error(y_teste, previsao_rln):,.2f}')

```

O melhor modelo √© o Arvore de Decis√£o.

**Passo 4 - Usar melhor modelo**

---

```python
# Usar o melhor modelo -> Modelo Arvore de Decis√£o

#Importar os novos avi√µes para previs√£o
df_novos_dados = pd.read_csv("car_price_novos.csv")

# Codificar o novo dataframe

# Codificar coluna Marca
codificador_marca = LabelEncoder()
df_novos_dados["Marca"] = codificador_marca.fit_transform(df_novos_dados["Marca"])

# Codificar coluna Tipo Modelo
codificador_modelo = LabelEncoder()
df_novos_dados["Modelo"] = codificador_modelo.fit_transform(df_novos_dados["Modelo"])

# Codificar coluna Tipo_Combustivel
codificador_combustivel = LabelEncoder()
df_novos_dados["Tipo_Combustivel"] = codificador_combustivel.fit_transform(df_novos_dados["Tipo_Combustivel"])

# Codificar coluna Transmissao
codificador_transmissao = LabelEncoder()
df_novos_dados["Transmissao"] = codificador_transmissao.fit_transform(df_novos_dados["Transmissao"])

# Visualizar o dataframe
display(df_novos_dados)

# Fazer a previs√£o atrav√©s do modelo √°rvore de decis√£o
nova_previsao = modelo_arvore_decisao.predict(df_novos_dados)

# Exibir a previs√£o
display(nova_previsao)
```
